### Cleaning and structuring the data

```python
# Load data
from pathlib import Path
csv_path = Path("data") / "transactions-2025-11-20.csv"
df = pd.read_csv(csv_path)


print("Raw shape:", df.shape)
print("\nColumns:", df.columns.tolist())
display(df.head())
```

Comment: Based on the summary above, several variables I think are not insignificant in explaining the transaction price ('TRANS_VALUE'). These are: "TRANSACTION_NUMBER", "INSTANCE_DATE", "TOTAL_BUYER", "TOTAL_SELLER", "MASTER_PROJECT_EN", "PROJECT_EN". Therefore, I will drop these columns for my new dataset. 

I will also need to encode variables such as bedrooms ('BEDROOMS'), group code ('GROUP_EN'), procedure code ('PROCEDURE_CAT'), Off-plan or not ('IS_OFFPLAN_CODE') and Free-hold or not ('IS_FREE_HOLD_CODE')

```python
# Helper functions for feature engineering
def extract_bedrooms(value):
    """
    Convert ROOM text like "3 B/R", "STUDIO", "2 BED" to a bedroom count.
    - STUDIO -> 0
    - first integer found in the string -> that number
    """
    if pd.isna(value):
        return np.nan
    s = str(value).upper().strip()
    if "STUDIO" in s:
        return 0
    m = re.search(r"\d+", s)
    if m:
        try:
            return int(m.group())
        except ValueError:
            return np.nan
    return np.nan

def map_procedure_category(text):
    """
    Simplify PROCEDURE_EN into a few categories.
    """
    if pd.isna(text):
        return "Other"
    t = str(text).lower()
    if "mortgage" in t:
        return "Mortgage Registration"
    elif "sale" in t or "transfer" in t:
        return "Sale Registration"
    elif "initial" in t:
        return "Initial Registration"
    elif "lease" in t:
        return "Lease Registration"
    elif "gift" in t:
        return "Gift Transfer"
    else:
        return "Other"
```

```python
# Basic cleaning & feature engineering

# Drop columns we don't need for modelling
cols_to_drop = [
    "TRANSACTION_NUMBER",
    "INSTANCE_DATE",
    "TOTAL_BUYER",
    "TOTAL_SELLER",
    "MASTER_PROJECT_EN",
    "PROJECT_EN",
]
df_clean = df.drop(columns=cols_to_drop, errors="ignore").copy()

# Bedrooms from ROOM text
df_clean["BEDROOMS"] = df_clean["ROOMS_EN"].apply(extract_bedrooms)
# Group code: simple categorical codes
if "GROUP_EN" in df_clean.columns:
    df_clean["GROUP_CODE"] = df_clean["GROUP_EN"].astype("category").cat.codes
else:
    df_clean["GROUP_CODE"] = 0  # fallback if missing

# Procedure code: map to a smaller set of categories, then to ints
df_clean["PROCEDURE_CAT"] = df_clean["PROCEDURE_EN"].apply(map_procedure_category)
procedure_map = {
    "Mortgage Registration": 1,
    "Sale Registration": 2,
    "Initial Registration": 3,
    "Lease Registration": 4,
    "Gift Transfer": 5,
    "Other": 99,
}
df_clean["PROCEDURE_CODE"] = df_clean["PROCEDURE_CAT"].map(procedure_map).astype(int)
```

```python
# IS_OFFPLAN_EN: Off-Plan = 1, everything else = 0
df_clean['IS_OFFPLAN_CODE'] = (
    df_clean['IS_OFFPLAN_EN']
      .astype(str).str.strip().str.lower()
      .map({'off-plan': 1, 'off plan': 1})
      .fillna(0)
      .astype(int)
)

# IS_FREE_HOLD_EN: Free Hold = 1, everything else = 0
df_clean['IS_FREE_HOLD_CODE'] = (
    df_clean['IS_FREE_HOLD_EN']
      .astype(str).str.strip().str.lower()
      .map({'free hold': 1, 'freehold': 1})
      .fillna(0)
      .astype(int)
)
```

### Feature engineering for models

```python
# One-hot encode AREA_EN and prepare X / y

# One-hot encode AREA_EN
df_model_dummies = pd.get_dummies(df_model, columns=["AREA_EN"], drop_first=True)

# Simple baseline features 
feature_cols_simple = ["PROCEDURE_AREA","ACTUAL_AREA", "BEDROOMS", 'PROCEDURE_CODE','IS_FREE_HOLD_CODE']
feature_cols_simple = [c for c in feature_cols_simple if c in df_model_dummies.columns]

# Extended features: baseline + other codes + AREA_EN dummies
extra_cols = ["IS_OFFPLAN_CODE", "GROUP_CODE"]
extra_cols = [c for c in extra_cols if c in df_model_dummies.columns]

area_dummy_cols = [c for c in df_model_dummies.columns if c.startswith("AREA_EN_")]

feature_cols_ext = feature_cols_simple + extra_cols + area_dummy_cols

print("Simple feature columns:", feature_cols_simple)
print("Extended feature columns:", feature_cols_ext)

X_ext = df_model_dummies[feature_cols_ext]
y = df_model_dummies["TRANS_VALUE"]
```

```python
# Single train/test split (all models will share this)
X_train_ext, X_test_ext, y_train, y_test = train_test_split(
    X_ext, y, test_size=0.2, random_state=42
)

# Derive "simple" feature matrices from the same split
X_train_simple = X_train_ext[feature_cols_simple]
X_test_simple = X_test_ext[feature_cols_simple]

print("Train shape (extended):", X_train_ext.shape)
print("Test shape  (extended):", X_test_ext.shape)
```

```python
# Helper functions
def evaluate_model(name, y_true, y_pred):
    """
    Print MAE, RMSE, R² and return them.
    """
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    print(f"{name:20s} -> MAE: {mae:,.0f} | RMSE: {rmse:,.0f} | R²: {r2:.3f}")
    return mae, rmse, r2

def plot_actual_vs_pred(y_true, y_pred, title):
    """
    Quick actual vs predicted scatter plot.
    """
    plt.figure(figsize=(6, 6))
    plt.scatter(y_true, y_pred, alpha=0.3)
    min_val = min(y_true.min(), y_pred.min())
    max_val = max(y_true.max(), y_pred.max())
    plt.plot([min_val, max_val], [min_val, max_val], "r--")
    plt.xlabel("Actual TRANS_VALUE")
    plt.ylabel("Predicted TRANS_VALUE")
    plt.title(title)
    plt.tight_layout()
    plt.show()
    
def plot_residual_diagnostics(y_true, y_pred, model_name, bins=50, hist_xlim=(-4_000_000, 4_000_000), alpha=0.2):
    y_true = np.asarray(y_true).ravel()
    y_pred = np.asarray(y_pred).ravel()
    residuals = y_true - y_pred

    # Histogram of residuals
    plt.figure(figsize=(8, 4))
    plt.hist(residuals, bins=bins)
    plt.title(f"Residuals (actual - predicted): {model_name}")
    plt.xlabel("Residual")
    plt.ylabel("Frequency")
    if hist_xlim is not None:
        plt.xlim(*hist_xlim)
    plt.tight_layout()
    plt.show()

    # Residuals vs predicted
    plt.figure(figsize=(8, 4))
    plt.scatter(y_pred, residuals, alpha=alpha)
    plt.axhline(0, color="black", linestyle="--")
    plt.title(f"Residuals vs Predicted: {model_name}")
    plt.xlabel("Predicted TRANS_VALUE")
    plt.ylabel("Residual (actual - predicted)")
    plt.tight_layout()
    plt.show()
```

### Modelling

```python
# Model 1 – Linear Regression (simple features)
lin_simple = LinearRegression()

t0 = time.perf_counter()
lin_simple.fit(X_train_simple, y_train)
fit_time = time.perf_counter() - t0

t0 = time.perf_counter()
y_pred_simple = lin_simple.predict(X_test_simple)
pred_time = time.perf_counter() - t0

mae_lin_simple, rmse_lin_simple, r2_lin_simple = evaluate_model(
    "Linear (simple)", y_test, y_pred_simple
)
plot_actual_vs_pred(y_test, y_pred_simple, "Linear (simple): Actual vs Predicted")

timings.append({
    "Model": "Linear Regression (Simple Features)",
    "Fit_s": fit_time,
    "Predict_s": pred_time,
    "MAE": mae_lin_simple,
    "RMSE": rmse_lin_simple,
    "R2": r2_lin_simple
})
```

```python
plot_residual_diagnostics(y_test, y_pred_simple, model_name="Simple")
```

```python
# Model 2 – Linear Regression (extended features)
lin_ext = LinearRegression()

t0 = time.perf_counter()
lin_ext.fit(X_train_ext, y_train)
fit_time = time.perf_counter() - t0

t0 = time.perf_counter()
y_pred_ext = lin_ext.predict(X_test_ext)
pred_time = time.perf_counter() - t0

mae_lin_ext, rmse_lin_ext, r2_lin_ext = evaluate_model(
    "Linear (extended)", y_test, y_pred_ext
)
plot_actual_vs_pred(y_test, y_pred_ext, "Linear (extended): Actual vs Predicted")

timings.append({
    "Model": "Linear Regression (Extended + Area Dummies)",
    "Fit_s": fit_time,
    "Predict_s": pred_time,
    "MAE": mae_lin_ext,
    "RMSE": rmse_lin_ext,
    "R2": r2_lin_ext
})
```

```python
plot_residual_diagnostics(y_test, y_pred_ext, model_name="Extended")
```

```python
# Model 3 – Random Forest
rf = RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    random_state=42,
    n_jobs=-1,
)

t0 = time.perf_counter()
rf.fit(X_train_ext, y_train)
fit_time = time.perf_counter() - t0

t0 = time.perf_counter()
y_pred_rf = rf.predict(X_test_ext)
pred_time = time.perf_counter() - t0

mae_rf, rmse_rf, r2_rf = evaluate_model("Random Forest", y_test, y_pred_rf)
plot_actual_vs_pred(y_test, y_pred_rf, "Random Forest: Actual vs Predicted")

timings.append({
    "Model": "Random Forest Regressor",
    "Fit_s": fit_time,
    "Predict_s": pred_time,
    "MAE": mae_rf,
    "RMSE": rmse_rf,
    "R2": r2_rf
})
```

```python
plot_residual_diagnostics(y_test, y_pred_rf, model_name="Random Forest")
```

```python
# Feature importances from Random Forest
importances = rf.feature_importances_
feature_names = X_train_ext.columns

feat_importances = pd.Series(importances, index=feature_names)

# Top 10 most important features
top_10 = feat_importances.sort_values(ascending=False).head(10)

plt.figure(figsize=(8, 4))
top_10.sort_values().plot(kind='barh')
plt.xlabel("Feature importance")
plt.title("Random Forest – Top 10 features")
plt.tight_layout()
plt.show()
```

```python
# Model 4 – Neural Network (simple MLP)

# Scale features for the neural network
scaler = StandardScaler()
X_train_dl = scaler.fit_transform(X_train_ext)
X_test_dl = scaler.transform(X_test_ext)

model_dl = keras.Sequential([
    layers.Input(shape=(X_train_dl.shape[1],)),
    layers.Dense(64, activation="relu"),
    layers.Dense(32, activation="relu"),
    layers.Dense(1),
])

model_dl.compile(optimizer="adam", loss="mse")

early_stop = keras.callbacks.EarlyStopping(
    monitor="val_loss", patience=5, restore_best_weights=True
)

t0 = time.perf_counter()
history = model_dl.fit(
    X_train_dl, y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=256,
    callbacks=[early_stop],
    verbose=0,
)
fit_time = time.perf_counter() - t0

t0 = time.perf_counter()
y_pred_dl = model_dl.predict(X_test_dl, verbose=0).ravel()
pred_time = time.perf_counter() - t0

mae_dl, rmse_dl, r2_dl = evaluate_model("Neural Net", y_test, y_pred_dl)
plot_actual_vs_pred(y_test, y_pred_dl, "Neural Net: Actual vs Predicted")

timings.append({
    "Model": "Neural Network (MLP)",
    "Fit_s": fit_time,
    "Predict_s": pred_time,
    "MAE": mae_dl,
    "RMSE": rmse_dl,
    "R2": r2_dl
})
```

```python
plot_residual_diagnostics(y_test, y_pred_dl, model_name= "Deep Learning")
```

```python
# Training curves (loss & MAE)
plt.figure(figsize=(8, 4))
plt.plot(history.history['loss'], label='Train loss')
plt.plot(history.history['val_loss'], label='Val loss')
plt.xlabel('Epoch')
plt.ylabel('MSE loss')
plt.title('Neural network training – loss')
plt.legend()
plt.tight_layout()
plt.show()
```

### Performance & limitations of four models

```python
def compare_models(models, name_map=None):
    name_map = name_map or {}
    rows = []
    for name, model, X_tr, y_tr, X_te, y_te, kind in models:
        pretty_name = name_map.get(name, name)
        pred = (lambda X: model.predict(X, verbose=0).ravel()) if kind == "keras" else model.predict
        yhat_tr = pred(X_tr)
        yhat_te = pred(X_te)
        r2_tr = r2_score(y_tr, yhat_tr)
        r2_te = r2_score(y_te, yhat_te)

        rows.append({
            "Model": pretty_name,
            "R² (Train)": r2_tr,
            "R² (Test)":  r2_te,
            "R² Gap":     r2_tr - r2_te,
            "RMSE (Train)": np.sqrt(mean_squared_error(y_tr, yhat_tr)),
            "RMSE (Test)":  np.sqrt(mean_squared_error(y_te, yhat_te)),
            "MAE (Train)":  mean_absolute_error(y_tr, yhat_tr),
            "MAE (Test)":   mean_absolute_error(y_te, yhat_te),
        })
    tbl = pd.DataFrame(rows)

    # formatting
    tbl[["R² (Train)", "R² (Test)", "R² Gap"]] = tbl[["R² (Train)", "R² (Test)", "R² Gap"]].round(4)
    for c in ["RMSE (Train)", "RMSE (Test)", "MAE (Train)", "MAE (Test)"]:
        tbl[c] = tbl[c].round(0).astype(int)

    # sort by best test R²
    tbl = tbl.sort_values("R² (Test)", ascending=False).reset_index(drop=True)    
    return tbl


# labels
name_map = {
    "Linear (simple)": "Linear Regression (Simple Features)",
    "Linear (extended)": "Linear Regression (Extended + Area Dummies)",
    "Random Forest": "Random Forest Regressor",
    "Neural Net": "Neural Network (MLP)",
}

models = [
    ("Linear (simple)",   lin_simple, X_train_simple, y_train, X_test_simple, y_test, "sklearn"),
    ("Linear (extended)", lin_ext,    X_train_ext,    y_train, X_test_ext,    y_test, "sklearn"),
    ("Random Forest",     rf,         X_train_ext,    y_train, X_test_ext,    y_test, "sklearn"),
    ("Neural Net",        model_dl,   X_train_dl,     y_train, X_test_dl,     y_test, "keras"),
]
```

```python
# 1) metrics table (train/test + gaps)
metrics_df = compare_models(models, name_map=name_map)

# 2) timing table (fit/predict/total + test metrics you stored)
timing_df = pd.DataFrame(timings).copy()
timing_df["Total_s"] = timing_df["Fit_s"] + timing_df["Predict_s"]

timing_df = timing_df[["Model", "Fit_s", "Predict_s", "Total_s"]].copy()
timing_df[["Fit_s", "Predict_s", "Total_s"]] = timing_df[["Fit_s", "Predict_s", "Total_s"]].round(4)

# 3) merge into ONE dataframe
final_df = metrics_df.merge(timing_df, on="Model", how="left")

# optional: reorder + format nicely
final_df = final_df[[
    "Model",
    "R² (Train)", "R² (Test)", "R² Gap",
    "RMSE (Train)", "RMSE (Test)",
    "MAE (Train)", "MAE (Test)",
    "Fit_s", "Predict_s", "Total_s"
]].copy()

# sort: best test R² first (change to "Total_s" if you want fastest first)
final_df = final_df.sort_values("R² (Test)", ascending=False).reset_index(drop=True)

final_df
```

Comments: From the results above, we can see that the NN model has the highest R-squared of 0.82, followed by the extended linear regression model (R-squared = 0.77), the random forest model (R-squared = 0.73), and the simple features linear regression model (R-squared = 0.52). 

However, we can also see that the R-squared gaps from each model are different; the simple features linear regression model has the lowest R-squared gap (0.0031), suggesting that it is the least overfitting model. Overall, the overfitting problem is not significant across the four models, as the dataset was split into 80% for training and 20% for testing before running each model.

Although the NN model has the highest R-squared, the total time it took to produce the results was 56 seconds, compared to less than 1 second for the linear regression models. The Random Forest model performs the worst in this project, with the longest total time of 62 seconds, the largest R-squared gap of 0.014 (indicating more overfitting than the other models), and an R-squared value of only 0.73.

### Rebuild a NN model for the 90% subset. 

```python
# Scale features for the neural net

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled   = scaler.transform(X_val)
X_arb_scaled   = scaler.transform(X_arb)


# Define & train Neural Network on training subset
input_dim = X_train_scaled.shape[1]
model_nn = keras.Sequential(
    [
        layers.Input(shape=(input_dim,)),
        layers.Dense(64, activation="relu"),
        layers.Dense(32, activation="relu"),
        layers.Dense(1),
    ]
)

model_nn.compile(optimizer="adam", loss="mse")

early_stop = keras.callbacks.EarlyStopping(
    monitor="val_loss",
    patience=5,
    restore_best_weights=True,
)

history = model_nn.fit(
    X_train_scaled,
    y_train,
    validation_data=(X_val_scaled, y_val),
    epochs=100,
    batch_size=256,
    callbacks=[early_stop],
    verbose=0,
)

# Validation metrics for the NN
y_val_pred = model_nn.predict(X_val_scaled).ravel()

print("\nNeural Net validation metrics:")
val_mae  = mean_absolute_error(y_val, y_val_pred)
val_rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
val_r2   = r2_score(y_val, y_val_pred)

print("MAE :", f"{val_mae:,.0f}")
print("RMSE:", f"{val_rmse:,.0f}")
print("R²  :", f"{val_r2:.3f}")
```

### Apply the NN model to identify arbitrage opportunities, using the 10% subset dataset.

```python
# Simple residual-based intervals on 10% ARBITRAGE set

# Residuals on validation set
residuals_val = (y_val - y_val_pred).to_numpy() if hasattr(y_val, "to_numpy") else np.asarray(y_val - y_val_pred)

# Take empirical error quantiles (e.g. 95% interval)
lower_err, upper_err = np.quantile(residuals_val, [0.025, 0.975])
print("Validation residual 2.5%:", f"{lower_err:,.0f}")
print("Validation residual 97.5%:", f"{upper_err:,.0f}")

# Base predictions on the arbitrage set
base_pred_arb = model_nn.predict(X_arb_scaled).ravel()

# Prediction intervals = base pred + residual quantiles
boot_mean = base_pred_arb
boot_low  = base_pred_arb + lower_err
boot_high = base_pred_arb + upper_err

# Build ARBITRAGE dataframe from original data (10% only)
arbitrage_df = df_clean.loc[X_arb.index].copy()
arbitrage_df["ACTUAL_PRICE"] = y_arb.values
arbitrage_df["PRED_MEAN"]    = boot_mean
arbitrage_df["PRED_LOW"]     = boot_low
arbitrage_df["PRED_HIGH"]    = boot_high

# Mispricing and % mispricing
arbitrage_df["MISPRICING"] = arbitrage_df["PRED_MEAN"] - arbitrage_df["ACTUAL_PRICE"]
arbitrage_df["MISPRICING_PCT"] = (
    arbitrage_df["MISPRICING"] / arbitrage_df["ACTUAL_PRICE"]
) * 100   # in %

# Simple arbitrage classification based on interval
def classify_arbitrage(row):
    if row["ACTUAL_PRICE"] < row["PRED_LOW"]:
        return "Undervalued"   # good buy candidate
    elif row["ACTUAL_PRICE"] > row["PRED_HIGH"]:
        return "Overvalued"
    else:
        return "Fairly priced"

arbitrage_df["ARBITRAGE_FLAG"] = arbitrage_df.apply(classify_arbitrage, axis=1)
```

##### Visualisation of the areas where opportunies exist

```python
# Check if the csv exist or not, if not then create a .csv file to capture the coordinates from the areas given in the dataset
coords_file = Path("area_coordinates.csv")

if coords_file.exists():
    area_coords_df = pd.read_csv(coords_file)
else:
    areas = (
        df_clean["AREA_EN"]
        .dropna()
        .astype(str)
        .str.strip()
        .unique()
    )

    geolocator = Nominatim(user_agent="dubai-arbitrage-mapper")
    geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)  # be nice to Nominatim

    results = []
    for area in areas:
        query = f"{area}, Dubai, United Arab Emirates"
        try:
            loc = geocode(query)
            results.append({
                "AREA_EN": area,
                "LATITUDE": getattr(loc, "latitude", np.nan),
                "LONGITUDE": getattr(loc, "longitude", np.nan),
            })
        except Exception as e:
            print(f"Error geocoding {area} -> {e}")
            results.append({"AREA_EN": area, "LATITUDE": np.nan, "LONGITUDE": np.nan})

    area_coords_df = pd.DataFrame(results)
    area_coords_df.to_csv(coords_file, index=False)

#  merge coordinates onto your dataframes
df_with_coords = df_clean.merge(area_coords_df, on="AREA_EN", how="left")

arbitrage_df_geo = arbitrage_df.merge(area_coords_df, on="AREA_EN", how="left")
arbitrage_df_geo = arbitrage_df_geo.dropna(subset=["LATITUDE", "LONGITUDE"]).copy()

print("Coords file:", coords_file)
print("area_coords_df shape:", area_coords_df.shape)
print("df_with_coords shape:", df_with_coords.shape)
print("arbitrage_df_geo shape:", arbitrage_df_geo.shape)
```
